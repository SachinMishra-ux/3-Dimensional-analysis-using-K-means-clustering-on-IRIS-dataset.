```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Let's lode the iris dataset.

```python
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
dir(iris)
```

```python
X=iris.data # all the data of four features.
X
```

```python
print(iris.feature_names)
```

## Lets create a dataframe using this iris data and target.

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

## Let's do the proper scaling of the features

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
scaler.fit(df[['petal length (cm)']])
df[['petal length (cm)']]=scaler.transform(df[['petal length (cm)']])
df
```

```python
scaler.fit(df[['petal width (cm)']])
df[['petal width (cm)']]=scaler.transform(df[['petal width (cm)']])
df
```

```python
scaler.fit(df[['sepal length (cm)']])
df[['sepal length (cm)']]=scaler.transform(df[['sepal length (cm)']])
df
```

```python
scaler.fit(df[['sepal width (cm)']])
df[['sepal width (cm)']]=scaler.transform(df[['sepal width (cm)']])
df
```

## Now let's run the k-means clustering and see the results.

```python
from sklearn.cluster import KMeans # K-means clustering algorithm
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]) # Runs the k-means clustering and gives clusters the data into 2 clusters.
y_predicted
```

```python
df['new_cluster']=y_predicted
df
```

```python
df1=df[df.new_cluster==0]
df2=df[df.new_cluster==1]
df3=df[df.new_cluster==2]
```

## Now let's plot the centroid of each cluster

```python
km.cluster_centers_
```

```python
bx = plt.axes(projection='3d') # This datapoints reprsents df1 features
bx.scatter3D(df1['sepal length (cm)'],df1['sepal width (cm)'],df1['petal length (cm)'],c=df1['petal width (cm)'], cmap='PuBuGn', alpha=0.4);
bx.scatter3D(df2['sepal length (cm)'],df2['sepal width (cm)'],df2['petal length (cm)'],c=df2['petal width (cm)'], cmap='PuBuGn', alpha=0.6);
bx.scatter3D(df3['sepal length (cm)'],df3['sepal width (cm)'],df3['petal length (cm)'],c=df3['petal width (cm)'], cmap='PuBuGn', alpha=0.6);
bx.scatter3D(km.cluster_centers_[:,0],km.cluster_centers_[:,1],km.cluster_centers_[:,2],km.cluster_centers_[:,3],marker='*',label='centroid',color='red',alpha=1)
plt.legend()
plt.figure(figsize=(10,7))
```
